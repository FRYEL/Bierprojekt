```python
corr = df.corr()

# 3) Erstelle den Plot
fig, ax = plt.subplots()
cax = ax.matshow(corr)                     # Heatmap
fig.colorbar(cax)                          # Farblegende
ticks = range(len(corr.columns))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns, rotation=90)
ax.set_yticklabels(corr.columns)
plt.title("Korrelationsmatrix")
#plt.tight_layout()
plt.show()
```

```python

# 2) Korrelationsmatrix berechnen
corr = df.corr()

# 3) Oberes Dreieck maskieren
mask = np.triu(np.ones_like(corr, dtype=bool))

# 4) Plot erstellen
fig, ax = plt.subplots(figsize=(16, 16))
cax = ax.matshow(corr, cmap='plasma', vmin=-1, vmax=1)
fig.colorbar(cax, fraction=0.046, pad=0.04)

# 5) Achseneinstellungen
ticks = np.arange(len(corr.columns))
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns, rotation=90, fontsize=8)
ax.set_yticklabels(corr.columns, fontsize=8)

# 6) Maske überzeichnen
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        if mask[i, j]:
            ax.add_patch(plt.Rectangle((j-0.5, i-0.5), 1, 1, color='white'))

plt.title("Korrelationsmatrix – Übersicht", pad=20, fontsize=14)
plt.tight_layout()
plt.show()
```
