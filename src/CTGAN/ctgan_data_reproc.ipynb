```python
import pandas as pd
from pyspark.sql.functions import column
```

```python
df = pd.read_csv('../../data/raw_data/original_dataset_schema.csv')
```

```python
df.describe()
```

```python
# Schritt 1: -99 in eigene Kategorie umcodieren (z. B. 4 oder 'unbekannt')
df['geschlecht'] = df['geschlecht'].replace(-99, 4)

# Optional: Als Kategorie umwandeln
df['geschlecht'] = df['geschlecht'].map({
    1: 'maennlich',
    2: 'weiblich',
    3: 'divers',
    4: 'unbekannt'
})
```

```python
df.describe()
```

```python
df['alter'] = df['alter'].replace(-99, 16)
```

```python
df['bundesland'] = df['bundesland'].replace(-99, 18)
```

```python
df['beruf'] = df['beruf'].replace(-99, 10)
```

```python
mapping = {
    1: "Angestellte / Angestellter",
    2: "Selbständige / Selbstständiger",
    3: "Studentin / Student",
    4: "Schülerin / Schüler",
    5: "Beamtin / Beamter",
    6: "Rentnerin / Rentner",
    7: "Hausfrau / Hausmann",
    8: "ohne Beschäftigung",
    9: "in Ausbildung/im Freiwilligendienst",
    10: "keine Angabe"
}

df['beruf'] = df['beruf'].map(mapping)
```

```python
mapping = {
    1: "täglich",
    2: "mehrmals pro Woche",
    3: "mehrmals im Monat",
    4: "einmal pro Monat",
    5: "seltener als einmal pro Monat"
}

df['konsumhaeufigkeit'] = df['konsumhaeufigkeit'].map(mapping)
```

```python
df['konsumhaeufigkeit'].value_counts()
```

```python
df.head(5)
```

```python
for col in df.select_dtypes(include='number').columns:
    median = df.loc[df[col] != -99, col].median()
    df[col] = df[col].replace(-99, median)
```
