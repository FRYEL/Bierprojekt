```python
missing_counts = (df == -99).sum()

# Nur Spalten anzeigen, in denen -99 vorkommen
missing_filtered = missing_counts[missing_counts > 0]

# Ausgabe anzeigen
print("Spalten mit -99-Werten:")
print(missing_filtered.sort_values(ascending=False))
```

```python
df.drop(columns=['bundesland'], inplace=True)
```

```python
df.info()
```

```python
df['geschlecht'].value_counts()
```

```python
# Schritt 1: -99 in eigene Kategorie umcodieren (z. B. 4 oder 'unbekannt')
df['geschlecht'] = df['geschlecht'].replace(-99, 4)

# Optional: Als Kategorie umwandeln
df['geschlecht'] = df['geschlecht'].map({
    1: 'maennlich',
    2: 'weiblich',
    3: 'divers',
    4: 'unbekannt'
})

# Schritt 2: One-hot encoding
df = pd.get_dummies(df, columns=['geschlecht'], prefix='geschlecht')
```

```python
# One-Hot-Encoding anwenden
df_encoded = pd.get_dummies(df, columns=[ 'beruf'], drop_first=True)
```

```python
df_encoded.columns
```

```python

# Liste aller Spalten (wie von dir angegeben)
columns = df_encoded.columns.tolist()

# Spalten, die **nicht** ordinal sind (bereits one-hot encoded → unangetastet!)
non_ordinal = [col for col in columns if col.startswith('geschlecht_') or col.startswith('beruf_')]

# Ordinale Spalten = alle außer den one-hot-Spalten
ordinal_cols = [col for col in columns if col not in non_ordinal]

# Schritt 1: -99 durch NaN ersetzen in den ordinalen Spalten
df_encoded[ordinal_cols] = df_encoded[ordinal_cols].replace(-99, np.nan)

# Schritt 2: NaN durch gerundete Spaltenmittelwerte ersetzen
df_encoded[ordinal_cols] = df_encoded[ordinal_cols].fillna(df_encoded[ordinal_cols].median().round())

# Optional: speichern

print("✅ Alle -99-Werte in ordinalen Spalten wurden durch Mittelwerte ersetzt. One-hot-Spalten blieben unverändert.")
```
